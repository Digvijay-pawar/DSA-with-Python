Implementation


```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = head
```

1) Traversal


```python
def traversalOfList(head):
  if head == None:
    return None
  print(head.data, end=" ")
  ptr = head.next
  while ptr != head:
    print(ptr.data, end=" ")
    ptr = ptr.next

traversalOfList(head)
```

2) Insert at beginning in O(n)

```python
h1 = Node(1)
h1.next = Node(2)
h1.next.next = h1
def insertAtStart(head, new):
  temp = Node(new)
  if head == None:
    temp.next = temp
    return temp
  ptr = head
  while ptr.next != head:
    ptr = ptr.next
  ptr.next = temp
  temp.next = head
  return temp
a = insertAtStart(h1, 0)
traversalOfList(a)
```

2) Insert at beginning in O(1)

```python
def insertAtStart1(head, new):
  temp = Node(new)
  if head == None:
    temp.next = temp
    return temp
  temp.next = head.next
  head.next = temp

  head.data, temp.data = temp.data , head.data
  return head
b = insertAtStart1(head, 12)
traversalOfList(b)
```

Delete start Node

```python
def deleteEnd(head):
  if head == None or head.next == None:
    return None
  ptr = head
  while ptr.next.next != head:
    ptr = ptr.next
  ptr.next = ptr.next.next
  return head

a = deleteEnd(head)
traversalOfList(a)
```
