**Traversing**

```python
class Node:
    def __init__(self, k):
        self.data = k
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

def print_list(head):
    curs = head
    while curs != None:
        print(curs.data, end=" ")
        curs = curs.next


print_list(head)
```

**Serching**

```python
def search_list(head, x):
    curs = head
    pos = 0
    while curs != None:
        pos += 1
        if curs.key == x:
            return pos
        else:
            curs = curs.next
    return -1
print(search_list(head, 30))
```

Insert at the beginning of Linked List

```python
class N:
    def __init__(self, key):
        self.key = key
        self.next = None

first = N(22)
first.next = N(33)
first.next.next = N(44)

print_list(first)
print()

def insert_start(head, key):
    temp = N(key)
    temp.next = head
    return temp

print_list(insert_start(first, 11))
```

**Insert at the End of the List**

```python
def insertEnd(head, data):
  temp = Node(data)
  if head == None:
    return temp
  ptr = head
  while ptr.next != None:
    ptr = ptr.next
  ptr.next = temp
  return head

print_list(first)
print_list(insertEnd(first, 10))

```

**Insert element at given position**

```python
def insertPos(head, pos, data):
  temp = Node(data)
  if pos == 1:
    temp.next  = head
    return temp
  ptr = head
  for i in range(pos - 2):
    ptr = ptr.next
    if ptr == None:
      return head
  temp.next = ptr.next
  ptr.next = temp
  return head

a = insertEnd(None, 10)
a = insertEnd(a, 20)
a = insertEnd(a, 40)
a = insertEnd(a, 50)
a = insertEnd(a, 60)

print_list(insertPos(a, 3, 30))
```

**Delete first Node of Linked List**

```python
def deleteFirst(head):
  if head == None:
    return None
  else:
    return head.next

b = deleteFirst(a)
print_list(b)
```

**Delete last Node of Linked List**

```python

def deleteLast(head):
  if head == None:
    return None
  if head.next == None:
    return None
  ptr = head
  while ptr.next.next != None:
    ptr = ptr.next
  ptr.next = None
  return head
c = insertEnd(None, 10)
c = insertEnd(c, 20)
print_list(c)
print_list(deleteLast(c))
```

Insert node in sorted linked list

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None


def insertSorted(head, x):
  temp = Node(x)
  if head == None:
    return temp
  elif x < head.data:
    temp.next = head
    return temp
  else:
    ptr = head
    while ptr.next != None and x > ptr.next.data:
      ptr = ptr.next
    temp.next = ptr.next
    ptr.next = temp
    return head

ab = Node(10)
ab.next = Node(20)
ab.next.next = Node(40)
ab.next.next.next = Node(50)
print_list(ab)
print()
print_list(insertSorted(ab, 45))
```

```python
def NthNode(head, n):
  if head == None:
    return None
  ptr = head
  len = 1
  while ptr.next:
    ptr = ptr.next
    len += 1
  x = (len - n) + 1
  ptr = head
  pos = 1
  if x < len:
    while x != pos:
      ptr = ptr.next
      pos += 1
    print(ptr.data)

NthNode(c, 5)
```
